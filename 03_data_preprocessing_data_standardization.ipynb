# Data Standardization & Variable Transformation

## Standardization

```python
import numpy as np
import pandas as pd

V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])

df = pd.DataFrame(
        {"V1" : V1,
         "V2" : V2,
         "V3" : V3}        
)



df = df.astype(float)
df
```

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)
```

this is a standardizetion with mean=0, std dev=1

## Normalization

transforming values of a variable into between 0-1; a kind of standardization

```python
preprocessing.normalize(df)
```

## Min-Max Transformation

transforming values of a variable into between any 2 numbers

```python
scaler = preprocessing.MinMaxScaler(feature_range = (10,20))
```

```python
scaler.fit_transform(df)
```

## Binarize Transformation

belirli bir eşik değerin altında veya üstündekileri 0 veya 1e dönüştürür; not standardization but transformation

```python
binarizer = preprocessing.Binarizer(threshold = 5).fit(df)
```

```python
binarizer.transform(df)
```

## 0-1 Transformation

transforming categorical variables into numeric variables 0 and 1

```python
import seaborn as sns
tips = sns.load_dataset('tips')
df = tips.copy()
df_l = df.copy()
```

```python
df_l.head()
```

```python
df_l["new_sex"] = df_l["sex"].cat.codes
```

```python
df_l.head()
```

```python
lbe = preprocessing.LabelEncoder()
```

```python
df_l["newer_sex"] = lbe.fit_transform(df_l["sex"])
```

```python
df_l.head()
```

## "1 and Others (0)" Transformation

transforming classes of categorical variable with more than 2 classes into "1" and "others"

```python
df.head()
```

```python
df_l.head()
```

```python
df_l["new_day"] = np.where(df_l["day"].str.contains("Sun"),1,0)
```

```python
df_l.head(20)
```

## Multiple Class Transformation

```python
lbe = preprocessing.LabelEncoder()
```

```python
df_l["newer_day"] = lbe.fit_transform(df_l["day"])
```

```python
df_l
```
